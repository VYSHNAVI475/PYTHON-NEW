##BASIC GRAPH

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([2, 8])
ypoints = np.array([1, 7])

plt.plot(xpoints, ypoints)
plt.show()
```

##BAR GRAPH

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["CLAD", "IT", "C", "ENGLISH","SINGLE","SEVERAL","GRAPHICS","PHYSICS","PRACTICALS","VDC"])
y = np.array([89, 97, 83, 85,89,82,97,72,94,76])

plt.bar(x,y)
plt.show()
```

##PIE CHART

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([88, 97, 80, 85,88,82,96,71,93,75])
mylabels = ["CLAD", "IT", "C", "ENGLISH","DISCRETE","GRAPHTHEORY","WORKSHOP","PHYSICS","PRACTICALS","VDC"]

plt.pie(y, labels = mylabels)
plt.show()
```

##GRAPH WITHOUT A LINE


```python
import matplotlib.pyplot as gv
import numpy as kb
a = kb.array([10,10])
m= kb.array([10,15])
gv.plot(a , m ,'o')
gv.show()
```

##MULTIPLE POINTS IN THE GRAPH

```python
import matplotlib.pyplot as gv
import numpy as kb
a = kb.array([1,3,5,7])
p = kb.array([2,4,6,8])
gv.plot(a,p)
gv.show()
```

##DEFAULT X POINTS

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([4, 9, 11, 1, 3, 5])

plt.plot(ypoints)
plt.show()
```

##MARKERS

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([4, 9, 11, 1])

plt.plot(ypoints, marker = 'x')
plt.show()
```

##DOTED GRAPH

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([4, 9, 11, 1])

plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

##DASHED

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([4, 9, 11, 1])

plt.plot(ypoints, linestyle = 'dashed')
plt.show()
```

##HISTOGRAMS
